# Importaciones generales

```python
import os
import pandas as pd
import requests
```

# Descargamos el DataSet de Meteoritos

```python
url = "https://data.nasa.gov/api/views/gh4g-9sfh/rows.csv?accessType=DOWNLOAD"
```

```python
ruta_carpeta = "/content/CSV's"

if not os.path.isdir(ruta_carpeta):
  os.mkdir(ruta_carpeta)
```

```python
nombre_arch = ruta_carpeta + '/meteoritos.csv'


if not os.path.isfile(nombre_arch):

  arch_csv = requests.get(url, stream=True)

  with open(nombre_arch,"wb") as csv:
      for chunk in arch_csv.iter_content(chunk_size=1024):
    
          if chunk:
              csv.write(chunk)
```

# Cargamos el DataSet

```python
df_meteoritos = pd.read_csv(nombre_arch)
```

## Variables Dummies

```python
#   Vemos Variables Dummies
pd.get_dummies(df_meteoritos['fall'])
```

```python
#   Creamos 2 nuevas columnas y convertimos variables de tipo String a variables de Tipo Numerico
df_meteoritos[['fell', 'found']] = pd.get_dummies(df_meteoritos['fall'])
df_meteoritos
```
