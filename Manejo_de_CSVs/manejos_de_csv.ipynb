```python
import os
import numpy as np
import pandas as pd
```

## Creando Carpeta

```python
ruta = '/content/csv\'s'
```

```python
if not os.path.isdir(ruta):
  os.mkdir(ruta)
```

```python
archivo = ruta + '/prueba.csv'
```

```python
dict_datos = {
    'edad': [10,9,13,14,12,11,12],
    'cm': [115,110,130,155,125,120,125],
    'pais':['co','mx','co','mx','mx','ch','ch'],
    'genero':['M','F','F','M','M','M','F'],
    'Q1':[5,10,8,np.nan,7,8,3],
    'Q2':[7,9,9,8,8,8,9]
}
```

## Guardando CSV's

```python
df = pd.DataFrame(dict_datos)
df
```

```python
df.to_csv(archivo, index=False)
```

## Leyendo CSV's

```python
df_read = pd.read_csv(archivo)
```

```python
df_read
```
